## 📝 学習詳細ノート

<details>
<summary><b>STEP 1. 水晶そのものを理解する</b></summary>

**🎯 目標**
* 「なぜ水晶で安定した発振ができるのか」を説明できるようになる。

**📖 学習内容**
* 直列共振と並列共振の違いを説明できる
* $C_0$ の役割を説明できる

**💡 学んだこと**
* 水晶には $L_m, C_m, R_m$ と $C_0$ からなる等価回路がある。
* $C_0$ があるため、直列共振と並列共振の2つの共振点が生じる。
    * **直列共振周波数**
      $$f_{s} = \frac{1}{2 \pi \sqrt{L_{m}C_{m}}}$$
    * **並列共振周波数**
      $$f_{p} = f_{s} \sqrt{1+\frac{C_{m}}{C_{0}}}$$
* 並列共振付近では水晶のインピーダンスが大きくなり、帰還回路の位相条件と利得条件が満たされやすくなる。
    * インピーダンスが最大だと、同じ $g_m$ に対して小信号電圧利得が最大となる。
    * 電源投入直後、回路には熱雑音の発生させる微弱なノイズしかない。
    * そのためバルクハウゼンの条件を満たすためには非常に大きなインピーダンスが必要。
* そのためその周波数が選択されて発振する。
* 水晶はQ値が非常に高いため周波数選択性が極めて鋭い。
* その結果、発振周波数がほぼ1点に固定され、非常に安定した発振が得られる。

</details>


<details>
<summary><b>STEP 2. 発振理論の基礎を、水晶発振器の文脈で学ぶ</b></summary>
**🎯 目標**
* 「どういう条件で発振し、なぜ起動するのか」を理解する。

**📖 学習内容**
* Barkhausen criterion (バルクハウゼンの条件)
* ループゲイン / Loop gain
* 位相条件 / Phase condition
* 起動条件 / Start-up condition
* 非線形振幅制限 / Nonlinear limiting
* 負性抵抗 / Negative resistance
* 励振レベル / Drive Level

**🚩 到達目標**
* Barkhausen条件と負性抵抗の関係を説明できる。
* 「発振する」と「確実に起動する」の違いを説明できる。
* 起動時と定常時で支配要因が異なることを理解する。

**💡 学んだこと**
* バルクハウンゼンの条件
    * ループゲイン$ A \beta \geq 1 $かつ位相条件$360°$であれば発振可能
* ただし発振可能だからと言って確実に起動できるわけではない。  
* 起動時は電源ノイズや熱雑音などの微小信号しか存在していない
* $ A \beta >= 1$を満たしていたとしても発振が成長するために非現実的な時間がかかってしまう
(あるいはPVTばらつきに耐えられない)
* 経験則から起動するためには$|R_{neg}| \geq 5 \times ESR$が必要だと言われている
* 上記条件を満たす場合、起動後振幅成長し増幅器のトランジスタが非線形動作となり増幅率が減少する
* 振幅は成長し続けるわけではなく振幅制限が起こり、しばらくすると定常状態になる
* まとめると、起動時は回路が水晶の損失以上のエネルギーを供給するため振幅が成長する。
* 振幅が大きくなると増幅器の利得あるいは負性抵抗が低下
* 最終的に$供給エネルギー=損失エネルギー$となる点、$A \beta = 1 (|R_{neg}| = ESR)$で安定

</details>

<summary><b>Step 3. Pierce発振器を徹底理解する</b></summary>

**🎯 目標**
* 最も基本的な水晶発振器の構成を、CMOS回路として説明・設計できるようにする。

**📖 学習内容**
* Pierce発振器 / Pierce oscillator
* CMOS inverter-based oscillator
* 帰還抵抗 / Feedback resistor
* 負荷容量 / Load capacitance
* バイアス点 / Bias point
* 起動性 / Start-up behavior

**🚩 到達目標**
* Pierce発振器の回路図を描ける
* 各素子（Rb, C1, C2, crystal）の役割を説明できる
* 発振周波数が何で決まるか説明できる

**💡 学んだこと**
    
* 帰還抵抗はAC信号を帰還しているのではなく、DC的なバイアス点を決めている
    * 自己バイアス  
    * インバータのゲートに電流はほとんど流れず、負荷容量にもDCは流れないので帰還抵抗がないと入力DCレベルが定まらない
    * インバータのgmが最も大きくなるインバータのしきい値付近で落ち着く
    
* ~~負荷容量C1, C2で反転増幅器によって180°位相が遅れた信号をさらに90°*2遅らせて入力へ帰還させている~~
    * 発振周波数は水晶と負荷容量のパラメータで決まる
    
        * 発振周波数$f_{OSC} \approx f_{s} * \sqrt{1+\frac{C_{m}}{C_{0}+C_{L}}}$
        * このとき
            * $f_s = \frac{1}{2\pi\sqrt{L_m C_m}}$
            * $C_L \approx \frac{C_1 C_2}{C_1 + C_2} + C_{stray}$

    * 負荷容量が付いたときは発振周波数が並列共振周波数から微妙にずれる
        * AC信号の帰還ネットワークは負荷容量+水晶(高Qな誘導性素子)で形成

* 負性抵抗の導出

    ![負性抵抗](./image/negR.png "negR")
    * $M_1$のゲートソース間に発生する電圧は$V_{GS} = -\frac{I_{X}}{C_{1}s}$
    * ドレイン電流は$I_{d} = -g_{m}*\frac{I_{X}}{C_{1}s}$
    * C2に流れる電流は$I_{X}-g_{m}*\frac{I_{X}}{C_{1}s}$
    * C2に発生する電圧は$\frac{(I_{X}-g_{m}*\frac{I_{X}}{C_{1}s})}{sC_{2}}$・・・①
    * C2に発生する電圧は$V_{X}+V_{GS} = V_{X} - \frac{I_{X}}{C_{1}s}$・・・②
    * ①、②より二つの端子から回路を見込んだインピーダンスは$Z = \frac{V_{X}}{I_{X}}(s) = \frac{1}{sC_{1}}+\frac{1}{sC_{2}}+\frac{g_{m}}{s^2 C_{1}C_{2}}$
    * $Z = \frac{V_{X}}{I_{X}}(j\omega) = \frac{1}{j \omega C_{1}}+\frac{1}{j \omega C_{2}}-\frac{g_{m}}{\omega^2 C_{1}C_{2}}$

* 起動特性と負性抵抗のトレードオフ

    * 負性抵抗を大きくするためには$C_{1}, C_{2}$を小さくしたい
    * ただし、発振周波数$f_{OSC} \approx f_{s} * \sqrt{1+\frac{C_{m}}{C_{0}+C_{L}}}$で決まるため$C_{1}, C_{2}$を小さくしすぎると、直列容量も小さくなり、$C_{stray}$のような寄生容量の影響が強く見えてしまう
    * 周波数精度と起動特性がトレードオフ

* 励振レベル

    * 起動後定常状態では$R_{neg}=ESR$となる
    * 水晶に投入される電力は$P_{drive} = {I_{x}}^2*ESR$
    * 起動余裕を確保しつつ励振レベルが大きくなり過ぎないようにすることが必要

</details>

<summary><b>Step 4. 負荷容量・寄生容量・周波数誤差を理解する</b></summary>  

**🎯 目標**
*「狙った周波数にならない理由」を説明できるようにする。

**📖 学習内容**
* Load capacitance (CL)
* Parasitic capacitance
* Pulling
* Frequency offset
* Package/Pad/ESD capacitance
* PCB parasitics

**🚩 到達目標**
* 実効CLの概念を説明できる
* パッド容量やESD容量が周波数に与える影響を説明できる
* 回路設計者として寄生をどう見積もるか整理できる

**💡 学んだこと**

* 負荷容量と発振周波数の関係
    * Pierce発振器の発振周波数は負荷共振周波数付近となり、
      $$f_{OSC} \approx f_s \sqrt{1+\frac{C_m}{C_0+C_L}}$$
      で近似できる。
    * このとき
      $$f_s=\frac{1}{2\pi\sqrt{L_mC_m}}$$
      である。
    * 周波数偏差は
      $$\frac{\Delta f}{f_s}=\frac{f_{OSC}-f_s}{f_s}=\sqrt{1+\frac{C_m}{C_0+C_L}}-1$$
    * $\frac{C_m}{C_0+C_L}\ll 1$なので、テイラー展開により
      $$\frac{\Delta f}{f_s}\approx \frac{C_m}{2(C_0+C_L)}$$
      となる。
    * 水晶の容量比を
      $$\gamma=\frac{C_0}{C_m}$$
      と定義すると、
      $$\frac{\Delta f}{f_s}\approx \frac{1}{2\gamma\left(1+\frac{C_L}{C_0}\right)}$$
      と表せる。
    * $\gamma$が大きいほど、負荷容量による周波数の引っ張り量は小さい。
    * 逆に$\gamma$が小さいほど、負荷容量による周波数感度は大きくなる。

* 実効負荷容量
    * 実際の負荷容量は理想的な$C_1, C_2$だけでは決まらない。
    * 水晶が見る実効負荷容量は、
      $$C_{L,eff}\approx \frac{(C_1+C_{p1})(C_2+C_{p2})}{(C_1+C_{p1})+(C_2+C_{p2})}$$
      と見積もれる。
    * ここで$C_{p1}, C_{p2}$には、ICパッド容量、ESD容量、パッケージ容量、PCB寄生容量、入力容量、出力容量などが含まれる。
    * 簡易的には
      $$C_{L,eff}\approx \frac{C_1C_2}{C_1+C_2}+C_{stray}$$
      と考える。

* 負荷容量と周波数ずれ
    * $C_{L,eff}$が大きいほど、発振周波数は低くなる。
    * $C_{L,eff}$が小さいほど、発振周波数は高くなる。
    * したがって、パッド容量、ESD容量、パッケージ容量、PCB容量を無視すると、狙った周波数からずれる。

* Pulling
    * Pullingとは、負荷容量の変化によって水晶の発振周波数が引っ張られる現象である。
    * 負荷容量感度が大きいほど、広い周波数調整範囲を得られる。
    * 一方で、容量ノイズや制御電圧ノイズも周波数変動に変換されやすくなる。

* チューニングレンジと位相ノイズ
    * 温度補償の幅、すなわちチューニングレンジを広げるためには、周波数感度を大きくする必要がある。
    * しかし感度を大きくすると、微小な容量変動や制御電圧ノイズが周波数変動として拡大される。
    * その結果、周波数ノイズが増え、位相ノイズやジッタが悪化しやすい。
    * したがって、チューニングレンジと位相ノイズはトレードオフになる。
